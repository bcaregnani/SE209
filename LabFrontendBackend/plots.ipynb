# Plot diferent graphs

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
unroll = np.array([1, 2, 4, 8, 16])
clkP = np.array([2780, 4050, 7800, 15000, 29500])
surface = np.array([1724, 4555, 11089, 26183, 56876])
cons = np.array([1.148e-03, 1.439e-03, 2.012e-03, 4.568e-03, 9.681e-03])
perf = np.array([11.2, 15.4, 16, 16.7, 17])
```

```python
# Plot performance vs unroll factor
plt.figure(figsize=(10, 5))
plt.plot(unroll, perf, 'o-')
plt.xlabel('Unroll Factor')
plt.ylabel('Performance (Mcypher/s)')
plt.title('Performance vs Unroll Factor')
plt.grid()
plt.savefig('perf_vs_unroll.png')
plt.show()
```

```python
# Plot clock period, surface, and energy consumption vs unroll factor in three subplots
fig, (ax1, ax2, ax3) = plt.subplots(3, 1)
ax1.set_xlabel('Unroll Factor')
ax1.set_ylabel('Clock Period (ps)')
ax1.plot(unroll, clkP, 'b.')
ax1.tick_params(axis='y')
ax1.set_xticks(unroll)
ax1.set_xticklabels(unroll)
ax1.set_title('Clock Period vs Unroll Factor')
ax1.grid()

ax2.set_xlabel('Unroll Factor')
ax2.set_ylabel('Surface (um^2)')
ax2.plot(unroll, surface, 'r.')
ax2.tick_params(axis='y')
ax2.set_xticks(unroll)
ax2.set_xticklabels(unroll)
ax2.set_title('Surface vs Unroll Factor')
ax2.grid()

ax3.set_xlabel('Unroll Factor')
ax3.set_ylabel('Energy Consumption (J)')
ax3.plot(unroll, cons, 'g.')
ax3.tick_params(axis='y')
ax3.set_xticks(unroll)
ax3.set_xticklabels(unroll)
ax3.set_title('Energy Consumption vs Unroll Factor')
ax3.grid()
```
